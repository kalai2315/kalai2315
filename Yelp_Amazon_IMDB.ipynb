````python
#Remove stopwords, punctuations, apply lematisation
````

````python
import string
````

````python
punct=string.punctuation
````

````python
punct
````

````python
from spacy.lang.en.stop_words import STOP_WORDS
````

````python
stopwords=list(STOP_WORDS)
````

````python
stopwords
````

````python
import spacy
nlp=spacy.load('en_core_web_sm')
````

````python
def text_cleaning(sentence):
  doc=nlp(sentence)
  tokens=[]
  for token in doc:
    if token.lemma_!="_PRON-":
      temp=token.lemma_.lower().strip()
    else:
      temp=token.lower_
    tokens.append(temp)

  cleaned_tokens=[]
  for token in tokens:
    if token not in stopwords and token not in punct:
      cleaned_tokens.append(token)
  return cleaned_tokens
````

````python
text_cleaning("In a word, it is embarrassing.		")
````

**Vectorisation Feature Engineering(TF-IDF)**


````python
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
````

````python
tfidf=TfidfVectorizer(tokenizer=text_cleaning)
#tokenisation will bw done according to this function

classifier=LinearSVC()
````

**Train the Model**

```
# This is formatted as code
```



````python
from sklearn.model_selection import train_test_split
````

````python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
````

````python
x_train.shape,x_test.shape
````

````python
x_train.head()
````

Fit the x_train and y_train

````python
clf=Pipeline([('tfidf',tfidf),('clf',classifier)])
````

````python
clf.fit(x_train,y_train)
````

Predict the Test Set Results

````python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
````

````python
y_pred=clf.predict(x_test)
````

````python
confusion_matrix(y_test,y_pred)
````

````python
print(classification_report(y_test,y_pred))
````

````python
accuracy_score(y_test,y_pred)
````
